```r
data = read.csv('data/parole.csv.xz')
summary(data)
str(data)
```

```r
table(data$violator)
```

```r
data$state = as.factor(data$state)
data$crime = as.factor(data$crime)
```

```r
summary(data)
```

```r
set.seed(144)
library(caTools)
split = sample.split(data$violator, SplitRatio=0.7)
train = subset(data, split==T)
test = subset(data, split==F)
```

```r
model = glm(violator ~ ., data = train, family='binomial')
summary(model)
```

```r
exp(1.489170)
```

```r
result = -4.2411574 + 1 * 0.3869904 + 1 * 0.8867192 -
    50 * 0.0001756 - 3 * 0.1238867 + 12 * 0.0802954 + 1 * 0.6837143
p_y_1 = 1 / (1 + exp(-result))
p_y_1
odds = p_y_1 / (1 - p_y_1)
odds
```

```r
pred = predict(model, newdata = test, type='response')
max(pred)
summary(pred)
```

```r
table(test$violator, pred >= 0.5)
```
